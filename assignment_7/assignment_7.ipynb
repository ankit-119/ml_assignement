```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#question 1
df=pd.read_csv('titanic.csv')
df.head()
#This is the weather history of canberra city
#This data shows date,location,MinTemp,Rainfall,Evaporation,Sunshine,WindGustDir,WindGustSpeed,WindDir,Pressure,Cloud,Temp,RainToday etc
```

```python
#question 2
print(df.shape)
df.dropna()
print(df.shape)
#Therefore is no missing values
```

```python
#question 3
plt.figure(0)
sns.boxplot(x=df['Age'])
plt.figure(1)
sns.boxplot(x=df['Fare'])
```

```python
#question 4
df['WithSomeone']=df['Siblings/Spouses Aboard']+df['Parents/Children Aboard']
plt.figure(0)
#this figure  shows graph of Pclass vs Fare graph 
sns.swarmplot(data=df,x='Pclass',y='Fare')
plt.figure(1)
#This figure show thne of people survived
sns.countplot(data=df,x='Survived')
plt.figure(2)
#This figure shows people in particular class
sns.countplot(data=df,x='Pclass')

plt.figure(3)
sns.scatterplot(data=df,x='Age',y='Fare')
plt.figure(4)
sns.catplot('WithSomeone',data=df,kind='count')
#this shows whether passenger was travelling alone or not
```

```python
#question 5
plt.figure(0)
sns.catplot('Pclass',data=df,hue='Sex',kind='count')
```
